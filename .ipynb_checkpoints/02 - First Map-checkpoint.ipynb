# Starter Libraries

```python
#GEOPandas, as introduced earlier
import geopandas as gpd

#Matplotlib, why not
import matplotlib.pyplot as plt
```

```python
#Local filepath for the data of interest
file_path = r'Data/COH_CENSUS_2010_TRACTS.shp'
gdf = gpd.read_file(file_path)
gdf.head()
```

# Your First Map

Is as easy as using `.plot()` on a fully loaded geopandas dataframe.  `.plot()` will take several arguments, all of which you can read about [here](https://geopandas.org/reference.html?highlight=plot#geopandas.GeoDataFrame.plot).

```python
gdf.plot(figsize = (10,10))
```

Easy enough! But that's not a lot of information. One of the first challenges in any analysis is defining a question. In this case, let's start with a simple spatial question, since all we have right now is spatial data:

**Where does the City of Houston have boundaries within Harris County? Does Harris County encompass all of Houston? Does Houston encompass all of Harris county?**

To do this, let's load in another file that exclusively describes the boundaries of the City of Houston.

```python
limits_file_path = r'Data\City_of_Houston_City_Limits__Full_and_Limited_Purpose_Areas_.shp'
limits = gpd.read_file(limits_file_path)
limits.head()
```

```python
limits.plot(figsize = (10,10))
```

So it looks like we've got spatial data on both our topics of interest. Let's put them both together and check on the results. Geopandas uses matplotlib under the hood, so, if we are going to superimpose multiple layers of data, we want to make sure we properly render them bottom to top in our code, as matplotlib draws in the specified order.

```python
#Geopandas works w/ the subplots method in matplotlib
fig, ax = plt.subplots(figsize = (10,10))
ax.set_title('Harris County vs City of Houston')

base = gdf.plot(ax = ax)
limits.plot(ax = base, color = 'orange')
```

# Hooray! It worked!

But we skipped a critical step. Remember we discussed in the last post that you want to make sure that you're using the same coordinate reference system before you join or superimpose two types of spatial data? Well, let's check the CRS of our dataframes using the `.crs` method on each.

```python
print(gdf.crs)
```

```python
print(limits.crs)
```

```python
gdf.crs == limits.crs
```

Phew! That's worked out!

# What if we wanted more information?

Another common extension available from commercial suites is `.dbf`. These contain all kinds of non-spatial data and directly associates it with spatial data. We can take a look at the `.dbf` files of our associated datasets right now to see what kind of information is included.

```python
#Note that we're storing the data in the same local directory, this is important
#The .dbf extension will also pull from the .prj extension and .shx extension to associate geometries and CRSs
limits_file_path = r'Data\City_of_Houston_City_Limits__Full_and_Limited_Purpose_Areas_.dbf'
limits = gpd.read_file(limits_file_path)
limits.head()
```

```python
file_path = r'Data/COH_CENSUS_2010_TRACTS.dbf'
tracts = gpd.read_file(file_path)
tracts.head()
```

Every column represents an attribute we can read and manipulate just like a standard Pandas dataframe. The `geometry` column at the tail end is unique to Geopandas, and it contains the associated geometry of the data we're looking at. In this case, the geometry column contains the geometry of each census district, and each column contains the associated data.

```python
#Attributes of census tract data
tracts.columns
```

# Spatial Data and Nonspatial Attributes

This association allows us to ask a variety of questions about the association between nonspatial attributes and spatial data. Specifically, we can look at the way nonspatial attributes are distributed within space. Let's say, for example, we want to ask a simple question:

**Where are the majority hispanic districts, relative to the COH's limits?**

We can answer this question first by manipulating our dataframe of census tracts. We're only intereted in those with a majority hispanic population, and we can pare down the list using the familiar `.loc` method.

```python
#Save this slice as a separate dataframe for plotting
hisp_majority = tracts.loc[(tracts['SUM_HispPo']/tracts['SUM_TotPop']) > 0.5]
hisp_majority.head()
```

```python
#Plot the hisp majority districts
hisp_majority.plot(figsize = (10,10))
```

Remember that our CRSs are the same on each frame? Well, they haven't changed. So, we're free to jump right into superposition.

```python
fig, ax = plt.subplots(figsize = (15,15))
ax.set_title('Majority Latino Districts vs COH Limits')

#Pink and purple are a good aesthetic
base = limits.plot(ax = ax, color = 'purple')
hisp_majority.plot(ax = base, color = 'pink', alpha = 0.7) #Setting the alpha < 0.7 makes it easier to read
```
